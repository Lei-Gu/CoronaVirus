```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
data=pd.read_csv("time_series-ncov-Confirmed.csv")
data.head()
```

```python
data_rename=data.rename(
        columns={"Province/State": "province_state",
                 "Country/Region": "country_region", 
                 "Lat": "latitude", 
                 "Long": "longitude", 
                 "Value": "confirmed_case",
                 "Date": "date"})
```

```python
data_rename.head()
```

```python
new_data=data_rename[
    ['country_region', 'province_state', 'latitude', 'longitude', 'date', 'confirmed_case']
].copy()
new_data.head()
```

```python
new_data.to_csv ('time_series-ncov-Confirmed.v2.csv', index = False, header=True)
```
