```python
global_population=pd.read_csv("Datasets/Global Population 2020.csv")
global_population.head()
```

```python
global_population.columns = global_population.iloc[0]
global_population = global_population.reindex(global_population.index.drop(0)).reset_index(drop=True)
global_population.columns.name = None
```

```python
global_population_rename=global_population.rename(
        columns={"Country (or dependency)": "country",
                 "Population (2020)": "population_2020",
                 "Med. Age": "med_age",
                 "World Share": "world_share"
                })
global_population_rename.head()
```

```python
global_population_updated = global_population_rename[['country','population_2020', 'med_age', 'world_share']]
```

```python
global_population_updated.head()
```

```python
global_population_updated['population_2020'] = global_population_updated['population_2020'].str.replace(',', '').astype(int)
```

```python
global_population_updated['med_age'] = global_population_updated["med_age"].str.replace('N.A.','0')
```

```python
global_population_updated['med_age']=global_population_updated['med_age'].astype(int)
```

```python
global_population_updated['world_share'] = global_population_updated['world_share'].str.rstrip('%').astype('float') / 100.0
```

```python
global_population_updated.dtypes
```

```python
global_population_updated.head()
```
